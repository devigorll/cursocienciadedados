```python
# Somando valores nulos por atributo
dataset.isnull().sum()
```

```python
# Mediana Salário
mediana_salario = sts.median(dataset["SALARIO"])

mediana_salario
```

```python
# Substituindo valores nulos pela mediana

dataset["SALARIO"].fillna(mediana_salario, inplace = True)
```

```python
# Verificando valores nulos em Gênero

dataset["GENERO"].isnull().sum()

agrupamento = dataset.groupby(["GENERO"]).size()
agrupamento
```

```python
dataset.loc[dataset["GENERO"] == "M", "GENERO"] = "Masculino"
```

```python
# Fazendo de forma individual

dataset.loc[dataset["GENERO"] == "F", "GENERO"] = "Feminino"
dataset.loc[dataset["GENERO"] == "Fem", "GENERO"] = "Feminino"

#Atualizando valores passando uma liista de valores que deverão ser substituidos

dataset.loc[dataset["GENERO"].isin(["F", "Fem"]), "GENERO"] = "Feminino"
```

```python
dataset["IDADE"].describe()
```

```python
dataset.loc[(dataset["IDADE"] < 0 ) | ( dataset["IDADE"] > 120 ) ]
```

```python
mediana_idade = sts.median(dataset["IDADE"])
mediana_idade
```

```python
dataset.loc[(dataset["IDADE"] < 0) | (dataset["IDADE"] > 120), "IDADE"] = mediana_idade

dataset["IDADE"].describe()
```

```python
# Retornando valores duplicados pelo ID

dataset[dataset.duplicated(["ID"], keep = False)]
```

```python
dataset.drop_duplicates(subset="ID", keep = "first", inplace = True)
dataset[dataset.duplicated(["ID"], keep = False)]
```

```python
agrupado = dataset.groupby(["ESTADO"]).size()

dataset.loc[dataset["ESTADO"].isin(["TD", "SP", "RP"]), "ESTADO"] = "RS"

agrupado
```

```python
# Calculando desvio padrão de salários para substituir ouliers

dp_salario = sts.stdev(dataset["SALARIO"])

dp_salario

dataset.loc[dataset["SALARIO"] > 2 * dp_salario]
```

```python
dataset.loc[dataset["SALARIO"] > 2 * dp_salario, "SALARIO"] = mediana_salario

dataset
```
